```python
aviation_df_clean['Weather.Condition'].isna().value_counts()
aviation_weather = aviation_df_clean['Weather.Condition'].dropna(inplace = False)
```

```python
aviation_df_clean['Broad.phase.of.flight'].isna().value_counts()
aviation_broad_phase_of_flight = aviation_df_clean['Broad.phase.of.flight'].dropna(inplace = False)
```

```python
aviation_df_clean['Total.Fatal.Injuries'].isna().value_counts()
aviation_total_fatal_injuries = aviation_df_clean['Total.Fatal.Injuries'].dropna(inplace = False)
```

```python
aviation_df_clean['Total.Uninjured'].isna().value_counts()
aviation_total_uninjured = aviation_df_clean['Total.Uninjured'].dropna(inplace = False)
```

Creating a new dataframe using the cleaned columns above.
'pd.DataFrame()' is used for creating the dataframe.
'.transpose()' is used to rotate the index and columns.

```python
COLUMNS = [aviation_broad_phase_of_flight, aviation_location, aviation_engine_type, aviation_injury_severity,aviation_make,aviation_weather,aviation_total_fatal_injuries,aviation_total_serious_injuries,aviation_total_minor_injuries,aviation_total_uninjured]
df_aviation = pd.DataFrame(data= COLUMNS, )
df_aviation_transpose= df_aviation.transpose()
df_aviation_transpose
```

Removing the missing values since not all columns are of the same lenght.

```python
df_aviation_transpose_cleaned = df_aviation_transpose.dropna(inplace= False)
df_aviation_transpose_cleaned
```

IMC - the visibility is too poor for the pilots to see.They rely on flight instruments for navigation.

VMC - The visibility is clear for the pilots. Opposite of IMC.

UNK - Unknown.

```python
weather_counts = df_aviation_transpose_cleaned['Weather.Condition'].value_counts()
weather_counts
```

```python
#Plotting a bar graph  to show the number of accidents in a certain weather condition
plt.figure(figsize = (12,7))
plt.bar(weather_counts.index, weather_counts)
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.title('Number of Accidents in Certain Weather Conditions')
plt.show()
```

Many accidents occurred on a clear day. Weather is a factor of the many accidents but it's not the main factor.
